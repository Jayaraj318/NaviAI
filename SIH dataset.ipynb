# 1. Flight Data from OpenSky Network (Using API)

```python
import requests
import pandas as pd

# OpenSky Network API URL for live flight data
opensky_url = "https://opensky-network.org/api/states/all"

# Send a request to the API
response = requests.get(opensky_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()

    # Extract relevant data fields
    flights = []
    for flight in data['states']:
        flights.append({
            'icao24': flight[0],
            'callsign': flight[1],
            'origin_country': flight[2],
            'time_position': flight[3],
            'last_contact': flight[4],
            'longitude': flight[5],
            'latitude': flight[6],
            'baro_altitude': flight[7],
            'on_ground': flight[8],
            'velocity': flight[9],
            'heading': flight[10],
            'vertical_rate': flight[11],
            'geo_altitude': flight[13],
            'squawk': flight[14],
            'spi': flight[15],
            'position_source': flight[16]
        })

    # Convert the data to a Pandas DataFrame
    df_flights = pd.DataFrame(flights)

    # Save to CSV
    csv_filename = 'flight_data.csv'
    df_flights.to_csv(csv_filename, index=False)

    print(f"Flight data collected and saved to {csv_filename}.")

    # Load the CSV file into a Pandas DataFrame
    df_flights_loaded = pd.read_csv(csv_filename)

    # Print the DataFrame to the console
    print("\nContents of the CSV file:")
    print(df_flights_loaded)
else:
    print(f"Failed to retrieve flight data. Status code: {response.status_code}")
```

```python
import requests
import pandas as pd

# OpenWeatherMap API key
api_key = "bd5e378503939ddaee76f12ad7a97608"

# List of cities to get weather data for
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix",
    "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose",
    "London", "Paris", "Berlin", "Madrid", "Rome",
    "Tokyo", "Beijing", "Shanghai", "Hong Kong", "Seoul",
    "Moscow", "Istanbul", "Dubai", "Mumbai", "Sydney",
    "Buenos Aires", "Rio de Janeiro", "Sao Paulo", "Lima", "Bogota",
    "Mexico City", "Toronto", "Vancouver", "Montreal", "Calgary",
    "Dubai", "Tel Aviv", "Riyadh", "Jeddah", "Doha",
    "Singapore", "Kuala Lumpur", "Bangkok", "Manila", "Ho Chi Minh City",
    "Jakarta", "Seoul", "Hanoi", "Melbourne", "Brisbane",
    "Auckland", "Wellington", "Lisbon", "Athens", "Vienna",
    "Zurich", "Geneva", "Stockholm", "Oslo", "Copenhagen",
    "Warsaw", "Budapest", "Prague", "Dublin", "Edinburgh",
    "Glasgow", "Barcelona", "Valencia", "Bilbao", "Seville",
    "Porto", "Helsinki", "Tallinn", "Riga", "Vilnius",
    "Lagos", "Cape Town", "Johannesburg", "Nairobi", "Accra",
    "Lima", "Santiago", "Buenos Aires", "Montevideo", "Asuncion",
    "Caracas", "Quito", "La Paz", "Sucre", "Santa Cruz",
    "San Salvador", "Guatemala City", "Tegucigalpa", "Managua", "Havanna",
    "Kingston", "Port-au-Prince", "Georgetown", "Paramaribo", "Bridgetown"
]

# OpenWeatherMap API URL for current weather
base_url = "https://api.openweathermap.org/data/2.5/weather?"

# List to store weather data for all cities
all_weather_data = []

for city_name in cities:
    # Complete URL with city name and API key
    complete_url = f"{base_url}q={city_name}&appid={api_key}"

    # Send a request to the OpenWeatherMap API
    try:
        response = requests.get(complete_url)

        # Check if the request was successful
        response.raise_for_status()

        # Parse the JSON data
        weather_data = response.json()

        # Flatten the JSON data
        flat_data = pd.json_normalize(weather_data)

        # Add the city name to the data for identification
        flat_data['city'] = city_name
        
        # Append the data to the list
        all_weather_data.append(flat_data)

    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred for {city_name}: {http_err}")
    except Exception as err:
        print(f"Other error occurred for {city_name}: {err}")

# Concatenate all the DataFrames into one
if all_weather_data:
    all_weather_df = pd.concat(all_weather_data, ignore_index=True)
    # Save the data to a CSV file
    csv_filename = "all_cities_weather_data.csv"
    all_weather_df.to_csv(csv_filename, index=False)
    print(f"Weather data for all cities saved to {csv_filename}.")

    # Load the CSV file into a Pandas DataFrame
    df_loaded = pd.read_csv(csv_filename)

    # Print the DataFrame to the console
    print("\nContents of the CSV file:")
    print(df_loaded)
else:
    print("No weather data was retrieved.")
```
