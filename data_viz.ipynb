```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<a id="1"></a>
# <p style="padding:10px;background-color:#27a102;margin:0;color:#111111;font-family:courier;font-size:100%;text-align:center;border-radius: 50px 50px;overflow:hidden;font-weight:bold">Explore the Data</p>

```python
df = pd.read_csv('shipment_data.csv')
df
```

```python
df.info()
```

```python
sns.set_style('white')
sns.set(rc={
    "axes.facecolor" : "#27a102",
    "figure.facecolor" : "#000000"

})
```

<a id="1"></a>
# <p style="padding:10px;background-color:#27a102;margin:0;color:#111111;font-family:courier;font-size:100%;text-align:center;border-radius: 50px 50px;overflow:hidden;font-weight:bold">Distribution of Barometric Altitude</p>

```python
sns.histplot(data=df, x='Barometric_Altitude', kde=True, bins=30, alpha=1, color='#ffbf00', edgecolor = 'white')
plt.xlim(0,16000)
plt.xlabel('Barometric Altitude').set_color('white')
plt.ylabel('Frequency').set_color('white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.show()
```

<a id="1"></a>
# <p style="padding:10px;background-color:#27a102;margin:0;color:#111111;font-family:courier;font-size:100%;text-align:center;border-radius: 50px 50px;overflow:hidden;font-weight:bold">Gepmetric Altitude</p>

```python
sns.histplot(data=df, x='Geometric_Altitude', kde=True, bins=30, alpha=1, color='#ffbf00' )
plt.xlim(0,16000)
plt.xlabel('Geometric Altitude').set_color('white')
plt.ylabel('Frequency').set_color('white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.show()
plt.show()
```

<a id="1"></a>
# <p style="padding:10px;background-color:#27a102;margin:0;color:#111111;font-family:courier;font-size:100%;text-align:center;border-radius: 50px 50px;overflow:hidden;font-weight:bold">Distribution of Heading</p>

```python
sns.histplot(data=df, x='Heading', kde = True, bins=30, alpha=1, color='#ffbf00', edgecolor = 'white', line_kws={'color': 'black'})
plt.xlabel('Heading').set_color('white')
plt.ylabel('Frequency').set_color('white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.show()
```

```python
sns.histplot(data=df, x='Velocity', kde=True, bins=30,alpha=1, color='#ffbf00', edgecolor = 'white' )
plt.xlim(0, 350)
plt.xlabel('Velocity').set_color('white')
plt.ylabel('Frequency').set_color('white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.show()
```
